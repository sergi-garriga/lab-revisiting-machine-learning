### Replace empty values with Nans

```python
df_cat.replace(' ', np.NaN, inplace=True)
df_cat.info()
```

```python
null_values = pd.DataFrame(df_cat.isna().sum(), columns=['null_values'])
null_values['total_values'] = 95412
null_values['%'] = round(null_values['null_values']/null_values['total_values']*100, 2)
null_values
```

### Features above 85% of null values

```python
features_to_drop = null_values[null_values['%'] > 85]
features_to_drop_list = list(features_to_drop.index)
features_to_drop_list
```

### Remove features

```python
df_cat.drop(columns=features_to_drop_list, axis=1, inplace=True)
```

### Dropped df

```python
dropped = df_cat.filter(['OSOURCE','ZIP'], axis=1)
dropped
```
